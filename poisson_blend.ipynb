```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.misc import imresize
import scipy
```

```python
def delete_rows_csr(mat, indices):
    """
    Remove the rows denoted by ``indices`` form the CSR sparse matrix ``mat``.
    """
    if not isinstance(mat, scipy.sparse.csr_matrix):
        raise ValueError("works only for CSR format -- use .tocsr() first")
    indices = list(indices)
    mask = np.ones(mat.shape[0], dtype=bool)
    mask[indices] = False
    return mat[mask]

def build_A_and_b(image):
    def b_vector(img):
        flattened = img.flatten()
        diff = np.diff(flattened)
        diff = np.append(diff, 0)
        diff = np.reshape(diff, img.shape)
        return diff[:, :-1].flatten()
    
    def A_matrix(image, row=True):
        num_row, num_col = image.shape
        total_elem = num_row*num_col
        if row:
            return -1 * delete_rows_csr(scipy.sparse.eye(total_elem) + scipy.sparse.diags([(total_elem-1)*[-1]], offsets=[1]), np.arange(num_col-1, total_elem, num_col))
        A = -1 * (scipy.sparse.eye(total_elem) + scipy.sparse.diags([(total_elem-num_col)*[-1]], offsets=[num_col]))[:-num_col, :]
        return A[np.array(list(np.arange(i, (num_row-1)*num_col, num_col) for i in range(num_col))).flatten(), :]
    A_row, b_row = A_matrix(image, row=True), b_vector(image)
    A_col, b_col = A_matrix(image, row=False), b_vector(np.transpose(image))
    A, b = scipy.sparse.vstack((A_row, A_col)), np.append(b_row, b_col)
    return A, b
```

```python
def naive_paste(source_image_in, target_image, target_region):
    target_image = target_image.astype(np.float32)
    t_x_min, t_y_min, t_x_max, t_y_max = target_region
    source_image_in = imresize(source_image_in, (t_y_max - t_y_min, t_x_max - t_x_min)).astype(np.float32)
    target_image[t_y_min:t_y_max, t_x_min:t_x_max] = source_image_in
    return target_image
```

```python
def poisson_blend(source_image_in, target_image, target_region):
    target_image = target_image.astype(np.float32)
    t_x_min, t_y_min, t_x_max, t_y_max = target_region
    source_image_in = imresize(source_image_in, (t_y_max - t_y_min, t_x_max - t_x_min)).astype(np.float32)
    target_image = target_image.astype(np.float32)
    for c in range(source_image_in.shape[2]):
        source_image = source_image_in[:, :, c]
        A, b = build_A_and_b(source_image)
        im2var = np.reshape(np.arange(np.prod(source_image.shape)), source_image.shape)
        
        for i in range(source_image.shape[0]):
            one_hot = np.zeros(np.prod(source_image.shape))
            one_hot[im2var[i, 0]] = 1
            A = scipy.sparse.vstack((A, one_hot))
            b = np.append(b, target_image[t_y_min + i, t_x_min - 1, c])
            
            one_hot = np.zeros(np.prod(source_image.shape))
            one_hot[im2var[i, source_image.shape[1] - 1]] = 1
            A = scipy.sparse.vstack((A, one_hot))
            b = np.append(b, target_image[t_y_min + i, t_x_max + 1, c])

        for i in range(source_image.shape[1]):
            one_hot = np.zeros(np.prod(source_image.shape))
            one_hot[im2var[0, i]] = 1
            A = scipy.sparse.vstack((A, one_hot))
            b = np.append(b, target_image[t_y_min - 1, t_x_min + i, c])
            
            one_hot = np.zeros(np.prod(source_image.shape))
            one_hot[im2var[source_image.shape[0] - 1, i]] = 1
            A = scipy.sparse.vstack((A, one_hot))
            b = np.append(b, target_image[t_y_max + 1, t_x_max + i, c])

        v = scipy.sparse.linalg.lsqr(A, b)[0]
        new_image = np.reshape(v, source_image.shape)
        target_image[t_y_min:t_y_max, t_x_min:t_x_max, c] = new_image
    return target_image
```

```python
penguin = np.array(Image.open('samples/penguin.jpg'))
```

```python
plt.imshow(penguin)
```

```python
source_penguin = penguin[190:450, 100:300]
plt.imshow(source_penguin)
```

```python
target = np.array(Image.open('samples/im3.jpg'))
plt.imshow(target)
```

# Naive Pasting

```python
result = naive_paste(source_penguin, target, [1000, 1500, 1500, 2000])
plt.imshow(result.astype(np.uint8))
```

# Poisson Blending

```python
result = poisson_blend(source_penguin, target, [1000, 1500, 1500, 2000])
plt.imshow(result.astype(np.uint8))
```
